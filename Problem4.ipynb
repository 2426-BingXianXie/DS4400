```python
import numpy as np


def problem_4_part_2():
    for i in range(1, 4):
        print(f"Matrix {i}")
        
        matrix_found = False
        while not matrix_found:
            A = np.random.randint(-10, 11, size=(3, 3))
            if np.linalg.det(A) != 0:
                matrix_found = True
        
        print("Generated Matrix:")
        print(A)
        
        A_inv = np.linalg.inv(A)
        print("\nInverse Matrix:")
        print(A_inv)
        
        identity_check = A @ A_inv
        
        print("\nProduct (Matrix * Inverse) - Identity Check:")
        print(np.round(identity_check, decimals=10))

if __name__ == "__main__":
    problem_4_part_2()
```
