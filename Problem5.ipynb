```python
import pandas as pd

# Load the dataset
df = pd.read_csv('train.csv')

# Drop the columns specified in the instructions
cols_to_ignore = ['id', 'date', 'zipcode']
# Use only columns that actually exist in the dataframe
existing_ignore = [c for c in cols_to_ignore if c in df.columns]
data = df.drop(columns=existing_ignore)

# 1. Compute Average, Min, Max, and Variance
stats = pd.DataFrame({
    'Average': data.mean(),
    'Min': data.min(),
    'Max': data.max(),
    'Variance': data.var()
})

print("Feature Statistics")
print(stats)

# Find Lowest/Highest Average
low_avg = stats['Average'].idxmin()
high_avg = stats['Average'].idxmax()

# Find Lowest/Highest Variance
low_var = stats['Average'].idxmin() # Variance often scales with mean; checking specifically
low_var = stats['Variance'].idxmin()
high_var = stats['Variance'].idxmax()

print(f"\nLowest Average: {low_avg} ({stats.loc[low_avg, 'Average']:.2f})")
print(f"Highest Average: {high_avg} ({stats.loc[high_avg, 'Average']:.2f})")
print(f"Lowest Variance: {low_var} ({stats.loc[low_var, 'Variance']:.2f})")
print(f"Highest Variance: {high_var} ({stats.loc[high_var, 'Variance']:.2f})")

# 2. Correlation with Response (price)
correlations = data.corr()['price'].sort_values(ascending=False)

print("\nCorrelation with Price")
print(correlations)

# Features with positive correlation
positive_corr = correlations[correlations > 0].index.tolist()
# Highest positive correlation (excluding price itself)
highest_pos = correlations.index[1] 

# 3. Negative correlations
negative_corr = correlations[correlations < 0]

print(f"\nPositively Correlated Features: {positive_corr}")
print(f"Feature with highest positive correlation: {highest_pos}")
print(f"Features with negative correlation:\n{negative_corr}")
```
